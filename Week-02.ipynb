#### Write a python program to draw a line using given axis values take from a text file in the x axis , y axis and title

```python
import matplotlib.pyplot as plt

x_label = 'x-axis'
y_label = 'y-axis'
title = 'Line plot of the values'

# Modified data
data = [(0, 0), (1, 2), (2, 4), (3, 6), (4, 8), (5, 10)]

x_values = [item[0] for item in data]
y_values = [item[1] for item in data]

plt.plot(x_values, y_values, label='Velocity', color='b')
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.legend()
plt.grid(True)  # Adding grid for better visualization
plt.show()
```

#### Create plot using subplots(), scatter, and some other methods on the Axes object:

```python
import matplotlib.pyplot as plt
import numpy as np


data_x, data_y = np.random.rand(50), np.random.rand(50)
bar_data = {'Category A': 20, 'Category B': 35, 'Category C': 15}

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Scatter plot
axs[0, 0].scatter(data_x, data_y, color='green', label='Scatter Plot')

# Line plot
axs[0, 1].plot(data_x, data_y, color='red', linestyle='--', marker='o', label='Line Plot')

# Bar chart
axs[1, 0].bar(bar_data.keys(), bar_data.values(), color='pink', label='Bar Chart', edgecolor='black', width=0.5)

# Histogram
axs[1, 1].hist(np.random.randn(1000), bins=30, color='purple', label='Histogram', edgecolor='black')

# Show the plot
plt.show()
```

### .Generate random data for x and y and create a scatter plot using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(70)
num_points = 60
x_values = np.random.rand(num_points) * 1.5
y_values = 3 * x_values + 2 + 0.45 * np.random.randn(num_points)

plt.scatter(x_values, y_values, label='Data', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Different Random Data')
plt.legend()
plt.show()
```

### Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own

```python
import matplotlib.pyplot as plt

categories = ['Housing', 'Utilities', 'Food', 'Entertainment', 'Transportation']
expenses = [1200, 250, 500, 150, 100]
myexplode = [0.02, 0.02, 0.02, 0.01, 0.01]

# Create a pie chart
plt.figure(figsize=(7, 8))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightgrey', 'gold', 'lightpink'], explode=myexplode)
plt.title('Expense Distribution in Budget')
plt.show()
```
