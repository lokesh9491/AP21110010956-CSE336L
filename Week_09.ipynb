```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.array([[2, 1], [3, 4], [5, 0], [7, 6], [9, 2]])
```

```python
centered_data = data - np.mean(data, axis=0)
cov_matrix = np.cov(centered_data, rowvar=False)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
sns.scatterplot(data)
plt.title("Original Domain Scatterplot")
plt.legend()
plt.grid()
```

```python

transformed_data = np.dot(centered_data, eigenvectors)
sns.scatterplot(transformed_data)
plt.title("Transformed Domain Scatterplot")
plt.legend()
plt.grid()




```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
data = iris.data
target = iris.target
mean = np.mean(data, axis=0)
normalized_data = data - mean
covariance_matrix = np.cov(normalized_data.T)

eig_val, eig_vec = np.linalg.eig(covariance_matrix)
print("Eigen Values: ",eig_val)
print("Eigen Vectors: ",eig_vec)
top = eig_vec[:, :2]
result = np.dot(normalized_data, top)
plt.figure(figsize=(8, 6))
for i in range(3):
    plt.scatter(result[target == i, 0], result[target == i, 1], label=f'Class {i}')
plt.title('PCA: Iris Dataset (2D Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
